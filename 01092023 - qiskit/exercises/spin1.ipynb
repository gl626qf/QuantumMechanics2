```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Rotation of a 1-dimensional vector around x axis using Gauss rotation

# Defining pauli_x matrix
pauli_x = np.array([[0, 1], [1, 0]])

# Defining the vector to be rotated
vector = np.array([[1], [0]])

# Defining the rotation angle
theta = np.pi / 1 * 2

# Defining the rotation matrix
# rotation_matrix = np.cos(theta/2)*np.identity(2) - 1j*np.sin(theta/2)*pauli_x

rotation_matrix = np.exp(-1j*pauli_x * theta/2)

# Rotating the vector
# rotated_vector = np.dot(rotation_matrix, vector)
rotated_vector = rotation_matrix @ vector

# Plotting the vector
# plt.plot(vector[0, 0], vector[1, 0], 'o', label='vector')
plt.plot([0, rotated_vector[1, 0]])
plt.plot([0, vector[0, 0]], [0, vector[1, 0]])
plt.xlabel('x')
plt.ylabel('y')
plt.axis('square')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.show()

```
