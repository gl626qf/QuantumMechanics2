```python
# Sakurai, J. J., & Napolitano, J. (2017). Modern quantum mechanics (2nd ed.).
# Cambridge: Cambridge University Press. doi:10.1017/9781316687857
# Problem 1.3 spin 1/2 


def inequality():

    deltaA = 0.5
    deltaB = 0.5
    A = np.array([[0, 1], [1, 0]])
    B = np.array([[0, -1j], [1j, 0]])

    RHS = expect(A, B) + deltaA * deltaB 
    return RHS




```

```python
quantumOperator = Qobj([[0, 1], [1, 0]])
```

```python
quantumState = Qobj([[1],[0]])
expect(quantumOperator, state = quantumState)
```

# Problem 1.3 
Evaluate inequality

```python
# Pauli spin matrices
sx = qutip.sigmax()
sy = qutip.sigmay()
sz = qutip.sigmaz()

# Spin up and spin down states
spin_up = qutip.basis(2,0)
spin_down = qutip.basis(2,1)

# Example: Calculating expectation values
# expectation_x = qutip.expect(sx,spin_up)
# expectation_y = qutip.expect(sy,spin_down)
# expectation_z = qutip.expect(sz, (spin_up + spin_down).unit())

```

```python
# Define initial state
initial_state = qutip.basis(2, 0)
# Define Hamiltonian
H = 0.5 * (sx + 0.5 * sy) 

#Time points for simulation
times = np.linspace(0.0, 10.0, 100)

#Perform simulation
result = qutip.mesolve(H, initial_state, times, [], [sx, sy, sz])


fig, axes = plt.subplots(1, 1)

axes.plot(times, result.expect[2], label=r"$\left<\sigma_z\right>$")
axes.plot(times, result.expect[1], label=r"$\left<\sigma_y\right>$")
axes.plot(times, result.expect[0], label=r"$\left<\sigma_x\right>$")

axes.set_xlabel(r"$t$", fontsize=20)
axes.legend(loc=2);
```
